```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df_1 = pd.read_csv('time_series_19-covid-Confirmed.csv')
df_2 = pd.read_csv('time_series_19-covid-Deaths.csv')
df_3 = pd.read_csv('time_series_19-covid-Recovered.csv')
```

```python
#dfs = ['df_1', 'df_2', 'df_3']
countries = ['Spain', 'Italy']
```

```python
df_1a = df_1[df_1['Country/Region'].isin(countries)] 
df_2a = df_2[df_2['Country/Region'].isin(countries)]
df_3a = df_3[df_3['Country/Region'].isin(countries)]
```

```python
df_1a
```

```python
df_1a.drop(['Province/State', 'Lat', 'Long'], axis=1, inplace=True)
df_2a.drop(['Province/State', 'Lat', 'Long'], axis=1, inplace=True)
df_3a.drop(['Province/State', 'Lat', 'Long'], axis=1, inplace=True)
```

```python
#for i in dfs:
df_1a['Status'] = 'Confirmed'
df_2a['Status'] = 'Dead'
df_3a['Status'] = 'Recovered'
```

```python
df = df_1a.append(df_2a)
df = df.append(df_3a)
```
